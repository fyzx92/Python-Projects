# Hnefatafl
It's an old norse board game with two asymmetric roles. The white pieces are all grouped in the middle, and are trying to protect their king to get him to one of the corners of the board, while the black pieces initially along the edges, and are trying to capture the king.

All pieces can move any distance in straight lines, but cannot jump over each other. For either side to capture a piece, they need to have two of their own pieces on opposite sides of the piece being captured. This does not apply to the king, he must be surrounded on all four sides to be captured.

The black pieces cannot go on to the corner squares, and if they are adjacent to a corner square, the white pieces can treat the corner square as an ally for the purposes of capturing.

First, we create the board, with indexed coordinates and safe squares.

Second, we need to be able to define pieces. Each piece has a position and color, and we need to know if the piece is the king or not. For the purposes of capturing, it is also useful to be able to compare the colors of two pieces.

```python
class Piece():
    """
    pieces to place on the board
    """
    
    def __init__(self, _pos, _color, _isking = False):
        """
        initiate piece position, color, king status
        """

        self.pos = _pos
        self.color = _color
        self.isking = _isking

    def is_samecolor(self, piece):
        """
        check if two pieces are on the same team
        """
        if self.color == piece.color:
            return True

        return False
```
